```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)

'''
Plotting different hydrodynamic properties vs refinement level, 
data extracted from mc_output_*.log files
'''

res_lev=[1,2,3,4,5]
r=[[8.907950e9,8.847328e9],[4.446371e9,4.431216e9],[2.225083e9,2.213717e9],
   [1.112067e9,1.107331e9],[5.561521e8,5.535472e8]]
theta=[[9.817500e-03,9.817500e-03],[4.908750e-03,4.908750e-03],[2.454375e-03,2.454375e-03],
   [1.227187e-03,1.227187e-03],[6.135937e-4,6.135937e-4]]
temp=[[5.346200e+06,5.321917e+06],[5.340135e+06,5.327993e+06],
      [5.343178e+06,5.324966e+06],[5.341659e+06,5.326482e+06],[5.342419e+06,5.325725e+06]]
gamma=[[1.000000e+02,1.000000e+02],[1.000000e+02,1.000000e+02],[1.000000e+02,1.000000e+02],
   [1.000000e+02,1.000000e+02],[1.000000e+02,1.000000e+02]]
density=[[1.759454e-07,1.735588e-07],[1.753472e-07,1.741539e-07],[1.756472e-07,1.738572e-07],
   [1.754975e-07,1.740058e-07],[1.755724e-07,1.739316e-07]]

dr=[]
dtheta=[]
dtemp=[]
dgamma=[]
ddens=[]

for i in range(len(r)):
    dr.append((r[i][0]+r[i][1])/2)

for i in range(len(theta)):
    dtheta.append((theta[i][0]+theta[i][1])/2)
    
for i in range(len(temp)):
    dtemp.append((temp[i][0]+temp[i][1])/2)   
    
for i in range(len(gamma)):
    dgamma.append((gamma[i][0]+gamma[i][1])/2)
    
for i in range(len(density)):
    ddens.append((density[i][0]+density[i][1])/2)
    
data=[dr,dtheta,dtemp,dgamma,ddens]
names=['radius','theta','temperature','gamma','density']

for i in range(len(data)):
    plt.scatter(res_lev,data[i], label='average {} for refinement level'.format(names[i]))
    #plt.yscale('log')
    plt.title("Refinement levels and {}".format(names[i]))
    plt.legend(loc="upper right")
    plt.xlabel("refinement levels")
    plt.ylabel("average {}".format(names[i]))
    plt.savefig("{}_lin.pdf".format(names[i]),dpi=600)
    plt.show()
    
    plt.scatter(res_lev,data[i], label='average {} for refinement level'.format(names[i]))
    plt.yscale('log')
    plt.title("Refinement levels and {}".format(names[i]))
    plt.legend(loc="upper right")
    plt.xlabel("refinement levels")
    plt.ylabel("average {}".format(names[i]))
    plt.savefig("{}_log.pdf".format(names[i]),dpi=600)
    plt.show()
```
