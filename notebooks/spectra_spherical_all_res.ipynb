We now define a blackbody function such that it outputs a blackbody spectrum to compare to our MCRaT output spectra.

```python
def blackbody_function(energies, temp, normalization, energy_unit=unit.keV):
	"""
	:param energies:
	:param temp:
	:param normalization:
	:param energy_unit:
	:return:
	"""

	energies=energies*energy_unit.to(unit.erg)
	try:
		energies=energies.value
	except AttributeError:
		energies = energies

	try:
		temp=temp.value
	except AttributeError:
		temp = temp
  
	try:
		normalization=normalization.value
	except AttributeError:
		normalization = normalization


	model =np.empty(energies.size)
	model=(energies**3/(const.h.cgs.value*const.c.cgs.value)**2)/(np.exp(energies/(const.k_B.cgs.value*temp))-1)
	energies = energies * unit.erg.to(energy_unit)
	model=model/np.trapz(model,x=energies)*normalization

	return model
```

```python
data=blackbody_function(spectrum_dict5_5['energy_bin_center'], 1.3e9, np.trapz(spectrum_dict5_5['spectrum'], x=spectrum_dict5_5['energy_bin_center']))

factor_x=(spectrum_dict5_5['energy_bin_center'][data.argmax()]/spectrum_dict5_5['energy_bin_center'][spectrum_dict5_5['spectrum'].argmax()])**-1
```
